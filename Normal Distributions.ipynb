```python
#Normal Distribution (mean-3,sigma=2)
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import seaborn as sns
mu, sigma = 3,2
s = np.random.normal(mu, sigma, 10000)
x = np.linspace(np.min(s),np.max(s))
plt.plot(x, norm.pdf(x,mu,sigma))
plt.hist(s, density=True)
plt.show()
```

```python
#sample normal distribtion (mean=1.5,sigma=2)
mu, sigma = 3,2
s = np.random.normal(mu, sigma, 10000)
x = np.linspace(np.min(s),np.max(s))
plt.plot(x, norm.pdf(x,mu,sigma), color = 'b')
plt.hist(s, density=True)

sample_mean, sample_sigma = 1.5, 2
sample = np.random.normal(sample_mean,sample_sigma, 200)
count, bins, ignored = plt.hist(s,30, alpha = 0.1, density = True)
y = np.linspace(np.min(sample),np.max(sample))
plt.plot(y, norm.pdf(y,sample_mean,sample_sigma), color='r')
plt.show()
```

```python
#finding confidence interval in sample distribution
confidenceInterval = norm.interval(alpha=.95,loc=sample_mean,scale=sample_sigma)
print(confidenceInterval)
```

```python
#added confidence intevals
mu, sigma = 3,2
s = np.random.normal(mu, sigma, 10000)
x = np.linspace(np.min(s),np.max(s))
plt.plot(x, norm.pdf(x,mu,sigma), color = 'b')
plt.hist(s, density=True)

sample_mean, sample_sigma = 1.5, 2
sample = np.random.normal(sample_mean,sample_sigma, 200)
count, bins, ignored = plt.hist(s,30, alpha = 0.1, density = True)
y = np.linspace(np.min(sample),np.max(sample))
plt.plot(y, norm.pdf(y,sample_mean,sample_sigma), color='r')

plt.axvline(-2.419927969080108, color='g')
plt.axvline(5.419927969080108, color='g')
plt.show
```

```python
#z score
z_score = (sample_mean - mu)/sigma
print(z_score)
```

```python
#finding p value, accept/reject null hypothesis
from statsmodels.stats import weightstats as stests
ztest ,pval = stests.ztest(s, x2=sample, value=0,alternative='two-sided')
print(float(pval))
if pval<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```
